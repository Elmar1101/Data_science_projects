```python
languages=['Python','C#','C++','Go']
popularity=[10,20,30,40]

plt.pie(popularity,labels=languages);
```

```python
languages=['Python','C#','C++','Go']
popularity=[10,20,30,40]

plt.pie(popularity,labels=languages, explode = [0,0,0.2,0], autopct='%.2f%%');
```

```python
fig, ax=plt.subplots(figsize=(10,7), nrows=2, ncols=2)

ax[0,0].pie(popularity, labels=languages, explode=[0,0,0.2,0],autopct='%.2f%%');
ax[0,1].bar(cars.keys(), cars.values());
ax[1,0].barh(list(cars.keys()), list(cars.values()));
ax[1,1].hist(random_data2,bins=40);
```

```python
plt.subplot(2,1,1)
plt.pie(popularity, labels=languages, explode=[0,0,0.2,0],autopct='%.2f%%');

plt.subplot(2,1,2)
plt.bar(cars.keys(), cars.values());
```
